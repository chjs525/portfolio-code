# 데이터 병합하기

```python
import pandas as pd
import os
import glob

# 파일 경로 패턴 설정 (현재 작업 디렉토리에 있을 경우)
file_pattern = 'CI_DMSTC_TOUR_AREA_INTRST_DGREE_INFO_2025*.csv'

# 해당 패턴에 맞는 모든 파일 경로 리스트 (파일명이 오름차순으로 정렬)
file_list = sorted(glob.glob(file_pattern))

# 파일별 데이터프레임을 읽고 리스트로 저장
df_list = [pd.read_csv(file) for file in file_list]

# 모든 주차 데이터를 하나의 데이터프레임으로 병합
df_interest_2025 = pd.concat(df_list, ignore_index=True)
```

## 지역별 관심도 값만 남기고 나머지 열 제거

```python
# 지역별 여행 관심도 컬럼만 추출
region_cols = [
    'SU_TOUR_INTRST_VALUE',     # 서울
    'GG_IC_TOUR_INTRST_VALUE',  # 경기인천
    'BS_TOUR_INTRST_VALUE',     # 부산
    'CHN_DJ_TOUR_INTRST_VALUE', # 충남대전
    'CB_TOUR_INTRST_VALUE',     # 충북
    'GN_US_TOUR_INTRST_VALUE',  # 경남울산
    'GB_DG_TOUR_INTRST_VALUE',  # 경북대구
    'JN_GJ_TOUR_INTRST_VALUE',  # 전남광주
    'JNBK_TOUR_INTRST_VALUE',   # 전북
    'GW_TOUR_INTRST_VALUE',     # 강원
    'JJ_TOUR_INTRST_VALUE'      # 제주
]

# 정제된 데이터프레임 이름: df
df = df_interest_2025[region_cols]
```

## 답변을 5점 척도로 변경

```python
# 전체 df에서 사용된 답변 유형 확인
unique_values = pd.unique(df.values.ravel())
print(unique_values)
print(f"총 응답 유형 개수: {len(unique_values)}")
```

- '많이 커졌다' -> 5
- '약간 커졌다' -> 4
- '예전과 비슷하다' -> 3
- '약간 적어졌다' -> 2
- '많이 적어졌다' -> 1

```python
# 문자 응답을 5점 척도로 매핑
scale_map = {
    '많이 커졌다': 5,
    '약간 커졌다': 4,
    '예전과 비슷하다': 3,
    '약간 적어졌다': 2,
    '많이 적어졌다': 1
}

# df 전체에 적용 (한 번에)
df = df.replace(scale_map)
```

```python
df
```

```python
# 1. 각 지역별 점수별 응답자 수 집계
score_counts = df.apply(pd.Series.value_counts).fillna(0).astype(int)

# 2. 지역이 행, 점수가 열인 형태로 보기 쉽게 정리
score_counts = score_counts.T  # 전치: 지역이 index, 점수가 column

# 3. 결과 확인
print(score_counts)
```

## 지역별 평균 관심도 점수 계산

```python
# ✅ 각 지역별 평균 관심도 점수 계산
# - df는 지역별 1~5점 관심도 데이터프레임
# - 각 지역(컬럼)별로 전체 응답자의 평균 점수를 계산
region_mean = df.mean()

# ✅ 평균 점수를 요약 데이터프레임 df1으로 변환
# - 컬럼명: 지역, 지역별_관심도
df1 = region_mean.reset_index()
df1.columns = ['지역', '지역별_관심도']
```

```python
df1
```

## 변수명 변경

```python
# 변수명 변경
df1.columns = ['지역', '지역별_관심도']
```

```python
df1
```

## 이중 지역인 것 분리

```python
# 복사본 생성
df2 = df1.copy()

# append 대신 concat을 위한 리스트 준비
rows_to_add = []

# 묶인 지역 → 분리 지역 딕셔너리
split_map = {
    '경기인천': ['경기', '인천'],
    '충남대전': ['충남', '대전'],
    '경남울산': ['경남', '울산'],
    '경북대구': ['경북', '대구'],
    '전남광주': ['전남', '광주']
}

# 각 묶인 지역 값 → 분리 지역으로 복사
for combined, splitted in split_map.items():
    value = df1.loc[df1['지역'] == combined, '지역별_관심도'].values[0]
    for region in splitted:
        rows_to_add.append({'지역': region, '지역별_관심도': value})

# 새 지역 행 추가
df2 = pd.concat([df1, pd.DataFrame(rows_to_add)], ignore_index=True)

# 기존 묶인 지역 제거 (← 여기 수정됨!)
df2 = df2[~df2['지역'].isin(split_map.keys())].reset_index(drop=True)
```

```python
df2
```

- 세종시는 통계 누락으로 인해 지리적 및 생활권 유사성이 큰 **대전시의 관심도 값(2.918471)**을 사용함

```python
# 1. 대전의 관심도 값을 가져옴
sejong_value = df2[df2['지역'] == '대전']['지역별_관심도'].values[0]

# 2. 세종 행을 df2에 추가함
df2.loc[len(df2)] = ['세종', sejong_value]
```

```python
df2
```

```python
# df2 저장 (CSV 형식, 인코딩 UTF-8)
df2.to_csv('지역별_관심도_최종.csv', index=False, encoding='utf-8-sig')
```
