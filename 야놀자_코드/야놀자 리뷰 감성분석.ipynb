## 모델 학습시 사용했던 정제내용 함수화 및 예측

```python
def sentiment_predict(new_sentence):
    new_sentence = re.sub(r'[^ㄱ-ㅎㅏ-ㅣ가-힣 ]','', new_sentence)
    new_sentence = okt.morphs(new_sentence)
    stopwords = ['도', '는', '다', '의', '가', '이', '은', '한', '에', '하', '고', '을', '를', '인', '듯', '과', '와', '네', '들', '듯', '지', '임', '게']
    new_sentence = [word for word in new_sentence if not word in stopwords]
    encoded = tokenizer.texts_to_sequences([new_sentence])
    pad_new = pad_sequences(encoded, maxlen=max_len)

    score = float(loaded_model.predict(pad_new))
    if score > 0.5:
        print("{:.2f}% 확률로 긍정 리뷰입니다.".format(score * 100))
    else:
        print("{:.2f}% 확률로 부정 리뷰입니다.".format((1 - score) * 100))
```

## 리뷰 감성분석: 점수화(0~100) 및 별점(1~5) 변환


```python
# === imports ===
import re
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.preprocessing.sequence import pad_sequences
from konlpy.tag import Okt
import pickle
from tqdm import tqdm  # 진행률 표시용

# tqdm을 pandas에 적용
tqdm.pandas()

# 전처리용 설정
max_len = 80 # 입력 문장 길이를 80 토큰으로 맞춤 (짧으면 패딩, 길면 자름)
okt = Okt()  # 한국어 형태소 분석기 (토큰화 용도)

# 모델과 토크나이저 로드
model_path = 'shopreviewnaver.keras'
tokenizer_path = 'shopreviewnaver.pkl'

loaded_model = tf.keras.models.load_model(model_path)

with open(tokenizer_path, 'rb') as handle:
    tokenizer = pickle.load(handle)

# 감성분석 결과를 0~100 범위의 정수 점수로 반환하는 함수
def sentiment_label(new_sentence):
    new_sentence = re.sub(r'[^ㄱ-ㅎㅏ-ㅣ가-힣 ]','', str(new_sentence))  # 문자열 아닌 경우 대비
    new_sentence = okt.morphs(new_sentence)
    stopwords = ['도', '는', '다', '의', '가', '이', '은', '한', '에', '하', '고', '을', '를', '인',
                 '듯', '과', '와', '네', '들', '듯', '지', '임', '게']
    new_sentence = [word for word in new_sentence if not word in stopwords]
    encoded = tokenizer.texts_to_sequences([new_sentence])
    pad_new = pad_sequences(encoded, maxlen=max_len)

    score = float(loaded_model.predict(pad_new, verbose=0))  # 0.0 ~ 1.0 확률 값
    return int(round(score * 100))  # 0~100 정수 변환

# === 전체 DataFrame(merged_df)에 적용 ===
# 리뷰내용 -> 0~100 감성점수
merged_df['감성점수(0-100)'] = merged_df['리뷰내용'].fillna('').progress_apply(sentiment_label)

# 0~100 -> 1~5 환산 함수
def divide_into_5(num: int) -> int:
    num = int(num)
    if not (0 <= num <= 100):
        raise ValueError("0~100 점수만 가능합니다.")
    return min(num // 20 + 1, 5)

# 최종 1~5 별점 
merged_df['감성별점(1-5)'] = merged_df['감성점수(0-100)'].apply(divide_into_5)

# 저장 (한글 파일명)
merged_df.to_csv('yanolja_reviews_감성점수.csv', index=False, encoding='utf-8-sig')
```

# 속도 개선한 감성분석 코드 (강사님 진행)

#### 배치(Batch) 예측

전체 데이터를 predict에 넣지만, 실제 연산은 배치 단위로 끊어서 진행된다
배치단위 : 128
예측 시간 : 약 30분

1. **속도 최적화** – GPU/CPU가 벡터 연산으로 한 번에 처리하므로 더 빠름. 병렬처리시 예측시간 2시간. 단 cpu 6코어 이상의 자원이 필요. 
2. **안정성** – Keras 모델이 병렬 스레드/프로세스에서 충돌할 위험이 없음.
3. **메모리 효율** – `batch_size`로 자동 분할 처리, 과부하 방지.
4. **코드 단순화** – 한 줄 `predict()` 호출로 전체 데이터 처리 가능.
5. **확장성** – 데이터가 커져도 배치 크기만 조절하면 쉽게 적용 가능.

## 모델, 토크나이저 로드

```python
import re
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.sequence import pad_sequences
from konlpy.tag import Okt
import pickle

max_len = 80
okt = Okt()
# 모델 파일 경로
model_path = 'shopreviewnaver.keras'
tokenizer_path = 'shopreviewnaver.pkl'


# 모델 불러오기
loaded_model = tf.keras.models.load_model(model_path)

with open(tokenizer_path, 'rb') as handle:
    tokenizer = pickle.load(handle)
```

## 리뷰 감성분석 진행: 별점(1~5) 변환

```python
from tqdm import tqdm
import re

# 전처리 함수 (단일 프로세스 버전)
def preprocess_text(text):
    clean_text = re.sub(r'[^ㄱ-ㅎㅏ-ㅣ가-힣 ]','', str(text))
    tokens = okt.morphs(clean_text)
    stopwords = ['도','는','다','의','가','이','은','한','에','하',
                 '고','을','를','인','듯','과','와','네','들','듯','지','임','게']
    tokens = [w for w in tokens if w not in stopwords]
    return tokens


def batch_sentiment_predict(df, text_col="내용", out_col="감성점수",
                            limit=None, batch_size=512, chunk_size=50000):
    if limit is not None:
        df = df.iloc[:limit].copy()
    else:
        df = df.copy()

    texts = df[text_col].tolist()

    # --- 전처리 (순차 실행: Windows 보안 충돌 없음) ---
    processed_texts = [preprocess_text(t) for t in tqdm(texts, desc="Preprocessing")]

    # --- 시퀀스 변환 ---
    encoded = tokenizer.texts_to_sequences(processed_texts)

    # --- Chunk 단위로 예측 ---
    scores_all = []
    for i in range(0, len(encoded), chunk_size):
        batch_encoded = encoded[i:i+chunk_size]
        pad_batch = pad_sequences(batch_encoded, maxlen=max_len)

        scores = loaded_model.predict(
            pad_batch, batch_size=batch_size, verbose=0
        ).flatten()

        scores_all.extend(scores)

    # --- 점수 매핑 ---
    def map_score(score):
        if score >= 0.90: return 5
        elif score >= 0.80: return 4
        elif score >= 0.40: return 3
        elif score >= 0.10: return 2
        else: return 1

    df[out_col] = [map_score(s) for s in scores_all]
    return df

yanol_ = batch_sentiment_predict(
    yanol, text_col="리뷰내용", out_col="감성별점", limit=None
)

yanol_
```
