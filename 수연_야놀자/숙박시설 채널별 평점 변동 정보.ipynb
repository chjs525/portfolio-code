## 기준일 변수 삭제 -> 월 단위로 합치기

[월 집계 결정] 
- 하루 단위 변동을 줄이고 비교·분석을 단순화하기 위해 월 단위로 평균·합계를 집계
- 집계 키: 숙박업명, 시도명, 구군명, 온라인여행사명, 기준년도, 기준월
- 평점값: 해당 월의 일별 평점 단순평균 → 단기 변동 완화 및 추세 파악 용이
- 리뷰등록수: 해당 월 리뷰 수 합계 → 월간 활동량 대표 지표 확보
- 기준일: 제거


```python
# [월 집계: df1을 월 단위로 덮어쓰기]
# - 키: 숙박업명, 시도명, 구군명, 온라인여행사명, 기준년도, 기준월
# - 평점값: 월 단순평균
# - 리뷰등록수: 월 합계
# - 기준일: 제거

keys = ['숙박업명','시도명','구군명','온라인여행사명','기준년도','기준월']

df1 = (
    df1
      .drop(columns=['기준일'], errors='ignore')            # 1) 일자 컬럼 제거
      .groupby(keys, as_index=False)                        # 2) 숙소/지역/OTA/연-월 단위 묶음
      .agg(평점값=('평점값','mean'),                        #    평점: 단순평균
           리뷰등록수=('리뷰등록수','sum'))                  #    리뷰수: 합계
      .assign(평점값=lambda d: d['평점값'].round(3))        # 3) 보기 좋게 평점 반올림(소수 3자리)
      .sort_values(keys, kind='stable')                     # 4) 키 기준 정렬
      .reset_index(drop=True)                               # 5) 인덱스 리셋
)
```

- 이런 상태에서 단순 평균·합계를 비교하면 샘플 수 차이로 인한 왜곡 발생

```python
# YANOLJA, GOODCHOICE가 있는 숙소만 추출
target_df = df1[df1['온라인여행사명'].isin(['YANOLJA', 'GOODCHOICE'])]

# 숙박업명으로 그룹화해서 두 여행사 모두 있는지 체크
both_exist = (
    target_df.groupby('숙박업명')['온라인여행사명']
    .nunique()
    .reset_index()
)

# 두 여행사 모두 있는 숙소만 필터
both_exist = both_exist[both_exist['온라인여행사명'] == 2]

# 전체 숙소 수
total_hotels = df1['숙박업명'].nunique()

# 개수와 비율 확인
count_both = len(both_exist)
ratio_both = count_both / total_hotels * 100

print(f"전체 숙소 수: {total_hotels}개")
print(f"두 여행사 모두 있는 숙소 개수: {count_both}개")
print(f"전체 숙소 대비 비율: {ratio_both:.2f}%")
```

```python
# YANOLJA, GOODCHOICE, AGODA가 있는 숙소만 추출
target_df = df1[df1['온라인여행사명'].isin(['YANOLJA', 'GOODCHOICE','AGODA'])]

# 숙박업명으로 그룹화해서 세 여행사 모두 있는지 체크
three_exist = (
    target_df.groupby('숙박업명')['온라인여행사명']
    .nunique()
    .reset_index()
)

# 세 여행사 모두 있는 숙소만 필터
three_exist = three_exist[three_exist['온라인여행사명'] == 3]

# 전체 숙소 수
total_hotels = df1['숙박업명'].nunique()

# 개수와 비율 확인
count_three = len(three_exist)
ratio_three = count_three / total_hotels * 100

print(f"전체 숙소 수: {total_hotels}개")
print(f"세 여행사 모두 있는 숙소 개수: {count_three}개")
print(f"전체 숙소 대비 비율: {ratio_three:.2f}%")
```

## 데이터 정제 (월 변수 삭제)

```python
keys = ['숙박업명','시도명','구군명','온라인여행사명']  # ① 집계 기준(숙소+지역+OTA)

df_월제거 = (
    df1.drop(columns=['기준월'])                       # ② 월 변수 제거(월은 안 씀)
       .groupby(keys, as_index=False)                 # ③ 같은 숙소+지역+OTA끼리 묶기
       .agg(평점값=('평점값','mean'),                 # ④ 평점: 평균
            리뷰등록수=('리뷰등록수','sum'))           #    리뷰: 합계
       .assign(평점값=lambda d: d['평점값'].round(3)) # ⑤ 평점 소수 3자리로 정리
       .sort_values(keys, kind='stable')              # ⑥ 키 기준 정렬
       .reset_index(drop=True)                        # ⑦ 인덱스 재설정
)
```

```python
# 동일 숙소 기준으로 야놀자 vs 여기어때 비교 (개수만)
keys = ['숙박업명','시도명','구군명']

# 두 OTA 모두 있는 숙소만
pair = df1[df1.groupby(keys)['온라인여행사명'].transform('nunique') == 2]

# 가로결합 (멀티인덱스 → 단일 컬럼명 변환)
wide = (
    pair.pivot(index=keys, columns='온라인여행사명', values=['평점값','리뷰등록수'])
        .dropna()
)
wide.columns = [f"{col1}_{col2}" for col1, col2 in wide.columns]
wide = wide.reset_index()

# 평점: 누가 더 높은지
cnt_rating_y = (wide['평점값_야놀자']   > wide['평점값_여기어때']).sum()
cnt_rating_g = (wide['평점값_여기어때'] > wide['평점값_야놀자']).sum()

# 리뷰개수: 누가 더 많은지
cnt_review_y = (wide['리뷰등록수_야놀자']   > wide['리뷰등록수_여기어때']).sum()
cnt_review_g = (wide['리뷰등록수_여기어때'] > wide['리뷰등록수_야놀자']).sum()

print(f"[평점] 야놀자 더 높음: {cnt_rating_y}개 / 여기어때 더 높음: {cnt_rating_g}개")
print(f"[리뷰] 야놀자 더 많음: {cnt_review_y}개 / 여기어때 더 많음: {cnt_review_g}개")
```

## 데이터 정제 (성수기·비성수기 구분)

```python
# df1 복사해서 새로운 데이터프레임 생성
df_시즌 = df1.copy()

# 7, 8, 12, 1월 = 성수기 / 나머지 = 비수기
peak_map = {7: '성수기', 8: '성수기', 12: '성수기', 1: '성수기'}
df_시즌['시즌구분'] = df_시즌['기준월'].apply(lambda x: peak_map.get(x, '비수기'))
```

```python
df_시즌
```

```python
# 1) 숙소·지역·OTA·성수기구분별 집계 (평점=평균, 리뷰=합계)
agg = (
    df_시즌
    .groupby(['숙박업명','시도명','구군명','온라인여행사명','시즌구분'], observed=True)
    .agg(평점값=('평점값','mean'),
         리뷰등록수=('리뷰등록수','sum'))
    .reset_index()
)

# 2) 성/비수기 피벗
pvt = (
    agg.pivot(index=['숙박업명','시도명','구군명','온라인여행사명'],
              columns='시즌구분',
              values=['평점값','리뷰등록수'])
       .reset_index()
)

# 3) 컬럼명 변경
pvt.columns = ['숙박업명','시도명','구군명','온라인여행사명',
               '비수기_평점','성수기_평점',
               '비수기_리뷰등록수','성수기_리뷰등록수']

# 4) 성수기·비수기 모두 있는 경우만
pvt = pvt.dropna(subset=['비수기_평점','성수기_평점','비수기_리뷰등록수','성수기_리뷰등록수'])

# 5) 같은 숙소·지역에 두 OTA 모두 있는 경우만
mask_two_ota = pvt.groupby(['숙박업명','시도명','구군명'])['온라인여행사명'].transform('nunique') == 2
df_시즌정제 = pvt[mask_two_ota].copy()

# 6) 값 정리
df_시즌정제['비수기_평점'] = df_시즌정제['비수기_평점'].round(3)
df_시즌정제['성수기_평점'] = df_시즌정제['성수기_평점'].round(3)
df_시즌정제['비수기_리뷰등록수'] = df_시즌정제['비수기_리뷰등록수'].astype(int)
df_시즌정제['성수기_리뷰등록수'] = df_시즌정제['성수기_리뷰등록수'].astype(int)
```

## 대응표본 분석 (동일 숙소 비교)

```python
from scipy import stats

# 1) 숙소 기준으로 두 DF 병합
merged = df_시즌_야놀자.merge(
    df_시즌_여기어때,
    on=['숙박업명', '시도명', '구군명'],
    suffixes=('_야놀자', '_여기어때')
)

# 2) 누가 더 높은지 개수 세기
metrics = ['비수기_평점', '성수기_평점', '비수기_리뷰등록수', '성수기_리뷰등록수']
count_results = {}

for m in metrics:
    y_count = (merged[f"{m}_야놀자"] > merged[f"{m}_여기어때"]).sum()
    g_count = (merged[f"{m}_야놀자"] < merged[f"{m}_여기어때"]).sum()
    count_results[m] = {'야놀자_우위': y_count, '여기어때_우위': g_count}

# 3) 대응표본 t-검정
ttest_results = {}
for m in metrics:
    t, p = stats.ttest_rel(merged[f"{m}_야놀자"], merged[f"{m}_여기어때"])
    ttest_results[m] = {'t값': t, 'p값': p}

# 4) 결과 출력
print("=== 개수 비교 ===")
for m, res in count_results.items():
    print(f"{m} → 야놀자 우위: {res['야놀자_우위']} / 여기어때 우위: {res['여기어때_우위']}")

print("\n=== 대응표본 t-검정 ===")
for m, res in ttest_results.items():
    print(f"{m} → t={res['t값']:.3f}, p={res['p값']:.4f}")
```

1. 개수 비교
- 평점: 비수기·성수기 모두 야놀자가 여기어때보다 더 높은 숙소 개수가 많음
- 비수기 리뷰등록수: 야놀자 우위 411개, 여기어때 우위 444개 → 여기어때가 약간 더 많은 숙소 보유
- 성수기 리뷰등록수: 야놀자 우위 264개, 여기어때 우위 590개 → 성수기 리뷰 수는 여기어때가 압도적으로 우위

2. 대응표본 t-검정 (평균 차이 검정)
- 비수기 평점: t=5.94, p<0.001 → 야놀자가 평균적으로 유의하게 높음
- 성수기 평점: t=2.73, p=0.0064 → 야놀자가 평균적으로 유의하게 높음
- 비수기 리뷰등록수: t=-2.63, p=0.0086 → 평균적으로 여기어때가 유의하게 많음
- 성수기 리뷰등록수: t=-6.62, p<0.001 → 평균적으로 여기어때가 유의하게 많음
